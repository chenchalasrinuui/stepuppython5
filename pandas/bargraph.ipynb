```python
from mysql.connector import connect
```

```python
con=connect(
 host="localhost",
 user="root",
 password="root",
 database="world"
)
```

```python
myCursor=con.cursor()
```

```python
query="select *from country"
```

```python
myCursor.execute(query)
```

```python
result=myCursor.fetchall()
```

```python
import pandas as pd
```

```python
columns=["Code","Name","Contient","Region","SurfaceArea","IndependentYear","Population","LifeExpectancy","GNP","GNP_old","LocalName","GovernmentForm","HeadOfState","Capital","Code2"]
```

```python
df=pd.DataFrame(result,columns=columns)
```

```python
df
```

```python
df.sort_values(by='Population', ascending=False,inplace=True)
```

```python
data=df.head()
```

```python
data
```
